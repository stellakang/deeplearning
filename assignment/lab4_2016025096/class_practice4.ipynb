```
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import random
import time
```

```
m=1000
n=100
epochs_num=1000
batch_sz=1000
```

```
x_train=np.zeros((m,2)) #x_train : 2 x m
y_train=np.zeros(m) #y_train : m x 1
x_test=np.zeros((n,2)) #x_test : 2 x n matrix
y_test=np.zeros(n) #y_test : n x 1 matrix
```

```
#generate m training samples
for i in range(m):
    x_train[i][0] = random.uniform(-2,2)
    x_train[i][1] = random.uniform(-2,2)
    if x_train[i][0] * x_train[i][0] > x_train[i][1]:
        y_train[i] = 1
    else:
        y_train[i] = 0
#generate n test samples
for j in range(n):    
    x_test[j][0] = random.uniform(-2,2)
    x_test[j][1] = random.uniform(-2,2)
    if x_test[j][0] * x_test[j][0] > x_test[j][1]:
        y_test[j] = 1
    else:
        y_test[j] = 0
```

```
with tf.device('/device:GPU:0'):
#with tf.device('/cpu:0'):
    model = keras.models.Sequential()
    model.add(layers.Input(shape=(2,)))
    model.add(layers.Dense(units=3, activation='sigmoid', 
                        use_bias=True,
                        kernel_initializer=keras.initializers.RandomNormal(), 
                        bias_initializer=keras.initializers.Zeros())
            )
    model.add(layers.Dense(units=1, activation='sigmoid',
                        use_bias=True,
                        kernel_initializer=keras.initializers.RandomNormal(),
                        bias_initializer=keras.initializers.RandomNormal())
            )
    model.compile(optimizer=keras.optimizers.SGD(learning_rate=3.7),
                loss=keras.losses.BinaryCrossentropy(),
                metrics=[keras.metrics.BinaryAccuracy()])
    start = time.time()
    model.fit(x_train, y_train, epochs=epochs_num, batch_size=batch_sz)
    print('train time: ',time.time()-start)
    start = time.time()
    model.evaluate(x_test, y_test, verbose=2)
    print('test time : ',time.time()-start)
```
